```python
# Simulate the JavaScript processing exactly
def simulate_js_processing():
    # Simulate reading from row 3 (range: 2 in pandas means skip first 2 rows)
    raw_data = df.iloc[2:].to_dict('records')
    
    def normalize_key(k):
        import re
        k = str(k).strip()
        k = re.sub(r'\s+', '_', k)
        k = re.sub(r'[^\w_]', '_', k)
        k = re.sub(r'__+', '_', k)
        k = re.sub(r'^_+|_+$', '', k)
        return k.lower()
    
    processed_data = []
    for row in raw_data:
        obj = {}
        # Normalize all keys
        for k, v in row.items():
            nk = normalize_key(k)
            obj[nk] = v if pd.notna(v) else ''
        
        # Apply convenience fields mapping
        obj['foto'] = obj.get('foto', '') or 'https://via.placeholder.com/120'
        obj['nis'] = obj.get('no_induk', '') or obj.get('nisn', '') or obj.get('nis', '') or obj.get('no', '') or ''
        obj['nama'] = obj.get('nama_lengkap', '') or obj.get('nama_panggilan', '') or obj.get('nama', '') or ''
        
        # Create TTL if not exists
        if not obj.get('ttl'):
            tempat = obj.get('tempat_lahir', '')
            tanggal = obj.get('tanggal_lahir', '')
            if tempat or tanggal:
                obj['ttl'] = f"{tempat}{', ' + str(tanggal) if tanggal else ''}".replace('^, ', '')
            else:
                obj['ttl'] = ''
        
        obj['alamat'] = obj.get('alamat_rumah_jalan', '') or obj.get('alamat_rumah', '') or obj.get('alamat', '') or ''
        obj['kelas'] = obj.get('kelas', '') or ''
        
        processed_data.append(obj)
    
    # Filter records with names (like JavaScript does)
    valid_students = [s for s in processed_data if s['nama']]
    
    return valid_students

students = simulate_js_processing()
print(f"Processed students count: {len(students)}")
print("\nSample students:")
for i, student in enumerate(students[:3]):
    print(f"Student {i+1}:")
    print(f"  NIS: {student['nis']}")
    print(f"  Nama: {student['nama']}")
    print(f"  TTL: {student['ttl']}")
    print(f"  Alamat: {student['alamat']}")
    print()

# Test search functionality
def test_search(keyword):
    keyword = keyword.lower()
    results = [s for s in students if 
               (s['nama'] and keyword in s['nama'].lower()) or 
               (s['nis'] and keyword in str(s['nis']).lower())]
    return results

print("Testing search for 'siswa':")
search_results = test_search('siswa')
print(f"Found {len(search_results)} results")
for result in search_results[:3]:
    print(f"- {result['nama']} (NIS: {result['nis']})")
```

```python
# Test the corrected JavaScript processing (reading from row 1 instead of row 2)
def simulate_corrected_js_processing():
    # Simulate reading from row 2 (range: 1 in pandas means skip first 1 row)
    raw_data = df.iloc[1:].to_dict('records')  # Skip only the first row (index 0)
    
    def normalize_key(k):
        import re
        k = str(k).strip()
        k = re.sub(r'\s+', '_', k)
        k = re.sub(r'[^\w_]', '_', k)
        k = re.sub(r'__+', '_', k)
        k = re.sub(r'^_+|_+$', '', k)
        return k.lower()
    
    processed_data = []
    for row in raw_data:
        obj = {}
        # Normalize all keys
        for k, v in row.items():
            nk = normalize_key(k)
            obj[nk] = v if pd.notna(v) else ''
        
        # Apply convenience fields mapping
        obj['foto'] = obj.get('foto', '') or 'https://via.placeholder.com/120'
        obj['nis'] = obj.get('no_induk', '') or obj.get('nisn', '') or obj.get('nis', '') or obj.get('no', '') or ''
        obj['nama'] = obj.get('nama_lengkap', '') or obj.get('nama_panggilan', '') or obj.get('nama', '') or ''
        
        # Create TTL if not exists
        if not obj.get('ttl'):
            tempat = obj.get('tempat_lahir', '')
            tanggal = obj.get('tanggal_lahir', '')
            if tempat or tanggal:
                obj['ttl'] = f"{tempat}{', ' + str(tanggal) if tanggal else ''}".replace('^, ', '')
            else:
                obj['ttl'] = ''
        
        obj['alamat'] = obj.get('alamat_rumah_jalan', '') or obj.get('alamat_rumah', '') or obj.get('alamat', '') or ''
        obj['kelas'] = obj.get('kelas', '') or ''
        
        processed_data.append(obj)
    
    # Filter records with names (like JavaScript does)
    valid_students = [s for s in processed_data if s['nama']]
    
    return valid_students

corrected_students = simulate_corrected_js_processing()
print(f"Corrected processing - Total students: {len(corrected_students)}")

# Test search for "nama siswa 1"
def test_search_corrected(keyword):
    keyword = keyword.lower()
    results = [s for s in corrected_students if 
               (s['nama'] and keyword in s['nama'].lower()) or 
               (s['nis'] and keyword in str(s['nis']).lower())]
    return results

print("\nTesting search for 'nama siswa 1':")
search_results = test_search_corrected('nama siswa 1')
print(f"Found {len(search_results)} results")
for result in search_results:
    print(f"- {result['nama']} (NIS: {result['nis']})")

print("\nFirst 5 students in corrected processing:")
for i, student in enumerate(corrected_students[:5]):
    print(f"{i+1}. {student['nama']} (NIS: {student['nis']})")
```
